```python
import pandas as pd

# --- Step 1: Load your weather data ---
weather_df = pd.read_csv("weather_data_clean.csv")

# Clean up: Fill missing timestamps with NaN, then drop those rows
weather_df = weather_df.dropna(subset=["timestamp"])

# Convert timestamp to date for merging
weather_df["date"] = pd.to_datetime(weather_df["timestamp"], errors="coerce").dt.date

# --- Step 2: Load/parse your NOAA rainfall JSON ---
# Replace this with your actual NOAA API JSON response
noaa_json = {
    "results": [
        {"date": "2025-09-16T00:00:00", "datatype": "PRCP", "station": "GHCND:US1NYNY0001", "value": 5},
        {"date": "2025-09-16T00:00:00", "datatype": "PRCP", "station": "GHCND:US1NJTJ0002", "value": 0},
        # ... more records ...
    ]
}

# Filter for rainfall (PRCP) only
rainfall = [r for r in noaa_json["results"] if r["datatype"] == "PRCP"]

rain_df = pd.DataFrame(rainfall)
rain_df["date"] = pd.to_datetime(rain_df["date"]).dt.date
rain_df = rain_df.rename(columns={"value": "rainfall"})

# Aggregate rainfall by date (sum for all stations)
rain_by_date = rain_df.groupby("date")["rainfall"].sum().reset_index()

# --- Step 3: Merge rainfall with weather data ---
merged = pd.merge(weather_df, rain_by_date, how="left", left_on="date", right_on="date")

# Optional: Drop the helper 'date' column
merged = merged.drop(columns=["date"])

# --- Step 4: Save the merged data ---
merged.to_csv("weather_data_with_rainfall.csv", index=False)

print("Merged data saved to weather_data_with_rainfall.csv")
```

```python
import pandas as pd

# --- Step 1: Load your weather data ---
weather_df = pd.read_csv("weather_data_clean.csv")

# Clean up: Fill missing timestamps with NaN, then drop those rows
weather_df = weather_df.dropna(subset=["timestamp"])

# Convert timestamp to date for merging
weather_df["date"] = pd.to_datetime(weather_df["timestamp"], errors="coerce").dt.date

# --- Step 2: Load/parse your NOAA rainfall JSON ---
# Replace this with your actual NOAA API JSON response
noaa_json = {
    "results": [
        {"date": "2025-09-16T00:00:00", "datatype": "PRCP", "station": "GHCND:US1NYNY0001", "value": 5},
        {"date": "2025-09-16T00:00:00", "datatype": "PRCP", "station": "GHCND:US1NJTJ0002", "value": 0},
        # ... more records ...
    ]
}

# Filter for rainfall (PRCP) only
rainfall = [r for r in noaa_json["results"] if r["datatype"] == "PRCP"]

rain_df = pd.DataFrame(rainfall)
rain_df["date"] = pd.to_datetime(rain_df["date"]).dt.date
rain_df = rain_df.rename(columns={"value": "rainfall"})

# Aggregate rainfall by date (sum for all stations)
rain_by_date = rain_df.groupby("date")["rainfall"].sum().reset_index()

# --- Step 3: Merge rainfall with weather data ---
merged = pd.merge(weather_df, rain_by_date, how="left", left_on="date", right_on="date")

# Optional: Drop the helper 'date' column
merged = merged.drop(columns=["date"])

# --- Step 4: Save the merged data ---
merged.to_csv("weather_data_with_rainfall.csv", index=False)

print("Merged data saved to weather_data_with_rainfall.csv")
```

```python
import pandas as pd
import os

# Load datasets
weather = pd.read_csv("weather_data_clean.csv")
station = pd.read_csv("Dataset_for_chatbot.csv")

# Prepare weather data: extract date and city
weather = weather.dropna(subset=["timestamp", "city"])
weather["date"] = pd.to_datetime(weather["timestamp"], errors="coerce").dt.date
weather = weather.dropna(subset=["date"])

# Prepare station data: extract date
station["date"] = pd.to_datetime(station["date"], errors="coerce").dt.date

# Map city names to coordinates (for merging)
city_coords = {
    "London": (51.5074, -0.1278),
    "Paris": (48.8566, 2.3522),
    "New York": (40.7128, -74.0060),
    "Tokyo": (35.6895, 139.6917),
    "Delhi": (28.6139, 77.2090)
}
# Add city column to station data by nearest coordinates
def find_city(lat, lon):
    for city, (clat, clon) in city_coords.items():
        if abs(lat - clat) < 2 and abs(lon - clon) < 2:
            return city
    return None
station["city"] = station.apply(lambda r: find_city(r["latitude"], r["longitude"]), axis=1)
station = station.dropna(subset=["city"])

# Merge on date and city
merged = pd.merge(
    station,
    weather[["date", "city", "temperature", "humidity", "weather", "wind_speed"]],
    on=["date", "city"],
    how="left"
)

# Optional: Add rainfall if you have it
if os.path.exists("weather_data_with_rainfall.csv"):
    rain = pd.read_csv("weather_data_with_rainfall.csv")
    rain["date"] = pd.to_datetime(rain["timestamp"], errors="coerce").dt.date
    rain = rain[["date", "city", "rainfall"]]
    merged = pd.merge(merged, rain, on=["date", "city"], how="left")
    merged["rainfall"] = merged["rainfall"].fillna(0)

# Clean up columns
final = merged.drop(columns=["station_id", "latitude", "longitude"])
final = final.dropna(subset=["anomaly"])  # Keep only rows with labels

# Save final dataset
final.to_csv("final_training_dataset.csv", index=False)

# Optionally, remove other CSVs
for f in [
    "weather_data_clean.csv",
    "weather_data_with_rainfall.csv",
    "weather_data.csv",
    "Dataset_for_chatbot.csv"
]:
    if os.path.exists(f):
        os.remove(f)

print("Final dataset saved as final_training_dataset.csv")
```
